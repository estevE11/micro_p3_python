<h2>Exercici 7</h2><br>
Un zoo determina el preu en funció de l’edat de la persona. Si la persona té 2 o menys anys entra gratis. Si l’edat està entre 3 i 12 inclosos, el preu és 14 €. Si té més de 12 i menys de 65 costa 23 € i pels jubilats costa 18 €. 
Es demana fer un programa que vagi llegint les edats de les persones d’un grup. Quan la persona que entra les edats escrigui ‘f’ es mostrarà el preu final que ha de pagar aquell grup.

```python
def calc_price(age):
  if age <= 2:
    return 0
  
  if age >= 3 and age <= 12:
    return 14

  if age < 65:
    return 23

  return 18

price = 0
ipt = int(input("Edat: "))
while ipt != "f":
  price += calc_price(int(ipt))
  ipt = input("Edat: ")

print("El preu total es de %d€" % (price))
```

<h2>Exercici 8</h2><br>
Al 1937, un matemàtic alemany anomenat Lothar Collatz va formular la següent hipòtesi (encara no provada) que diu així:<br>
1. agafa un nombre enter positiu c0<br>
2. Si és parell fes c0 = c0 /2<br>
3. en cas contrari, si és imparell c0 = c0 x3 + 1<br>
4. si c0 és diferent a 1, torna al punt 2<br>
La hipòtesi diu que independentment del nombre inicial, les iteracions convergeixen en el nombre 1.
Fes un programa que comprovi el que diu la hipòtesi, de manera que donat un nombre ens vagi calculant els valors intermedis obtinguts fins arribar a 1 i el nombre de passos necessàries.<br>
<u>Exemple:</u><br>
Input: 15<br>
Output:<br>
46,23,70,35,106,53,160,80,40,20,10,5,16,8,4,2,1<br>
Passos: 17

```python
c0 = int(input("Nombre enter: "))
steps = 1
print(str(c0), end=",")

while c0 != 1:
  if c0 % 2 == 0:
    c0 = c0/2
  else:
    c0 = c0 * 3 + 1
  steps += 1
  print(str(c0), end=",")
print("\nPassos: " + str(steps))
```

<h2>Exercici 9</h2><br>
Es vol que donat un nombre de blocs per construir una piràmide s'indiqui l'alçada que tindrà la mateixa.
Així per exemple si la piràmide tingués la forma següent:<br>
 <br> &nbsp&nbsp&nbsp&nbsp  *
 <br>  &nbsp&nbsp *&nbsp * 
<br>  &nbsp *&nbsp * &nbsp* 
<br>  El nombre de blocs és 6 i l'alçada 3
<br> <u>Exemples:</u>
<br> Input. Nombre de blocs: 6
<br> Output. alçada: 3<br>
----------------------------
<br>  Input.Nombre de blocs: 20
<br>  Output. Alçada: 5<br>
----------------------------
<br>  Input.Nombre de blocs 1000
<br>  Output. Alçada: 44<br>
----------------------------
<br>  Input. Nombre de blocs 2
 <br> Output. Alçada: 1

```python
n = int(input("Nombre de blocs: "))

def blocks(blocks):
    height = 0
    cont = 1
    while cont <= blocks:
        height += 1
        blocks -= cont
        cont +=1
    return height
print(blocks(n))
```
